```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
both_wine = pd.read_pickle('both_wine.pkl')
```

```python
both_wine_feat = list(both_wine)
```

```python
wines = both_wine.as_matrix()
```

```python
k = 2
estimators = {'kmeans_2_20': KMeans(n_clusters=2,n_init=20,random_state=5),
              'kmeans_2_30': KMeans(n_clusters=2,n_init=30,random_state=5),
              'kmeans_3_20': KMeans(n_clusters=3,n_init=20,random_state=5)}
```

```python
fignum = 1
for name, est in estimators.items():
    fig = plt.figure(fignum, figsize=(4,3))
    plt.clf()
    ax = Axes3D(fig, rect=[1, 1, 2, 2], elev=10, azim=135)
    
    plt.cla()
    est.fit(wines)
    labels = est.labels_
    centroids = est.cluster_centers_
    
    ax.scatter(wines[:,0],wines[:,1],wines[:,2],c=labels.astype(np.float))
    
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel('fixed acidity')
    ax.set_ylabel('volatile acidity')
    ax.set_zlabel('citric acid')
    
    ax.scatter(wines[:,3],wines[:,4],wines[:,5],c=labels.astype(np.float))
    
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel('residual sugar')
    ax.set_ylabel('chlorides')
    ax.set_zlabel('free sulfur dioxide')
    
    fignum = fignum + 1
```
