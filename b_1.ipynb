```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
both_wine = pd.read_pickle('both_wine.pkl')
```

```python
both_wine_feat = list(both_wine)
```

```python
wines = both_wine.as_matrix()
```

```python
k = 2
kmeans = KMeans(n_clusters=2,random_state=0).fit(wines)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
for i in range(k):
    obs = wines[np.where(labels==i)]
    plt.plot(obs[:,8],obs[:,1],'o')
    lines = plt.plot(centroids[i,0],centroids[i,1],'kx')
    plt.setp(lines,ms=2.0)
    plt.setp(lines,mew=2.0)
```
